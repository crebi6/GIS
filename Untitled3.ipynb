File Paths

```python
boundary_shapefile = '/content/drive/MyDrive/Kipkabus/Kaptagat MergeFINAL.shp'
raster_file = '/content/drive/MyDrive/kaptagat/Kaptagat_Forest_Station_S2_2017.tif'
output_file = '/content/drive/MyDrive/output/2017_output.tif'
clipped_output_file = '/content/drive/MyDrive/output/2017clipped_output.tif'
```

Loading Boundary Shapefile

```python
boundary = gpd.read_file(boundary_shapefile)
```

Loading Raster Data and Reprojecting

```python
with rasterio.open(raster_file) as src:
    dst_crs = boundary.crs.to_string()
    transform, width, height = calculate_default_transform(src.crs, dst_crs, src.width, src.height, *src.bounds)
    kwargs = src.meta.copy()
    kwargs.update({
        'crs': dst_crs,
        'transform': transform,
        'width': width,
        'height': height
    })

    with rasterio.open(output_file, 'w', **kwargs) as dst:
        for i in range(1, src.count + 1):
            reproject(
                source=rasterio.band(src, i),
                destination=rasterio.band(dst, i),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=transform,
                dst_crs=dst_crs,
                resampling=Resampling.nearest
            )
```

Clipping the Raster Using the Boundary Shapefile

```python
with rasterio.open(output_file) as src:
    out_image, out_transform = mask(src, boundary.geometry, crop=True)
    out_meta = src.meta.copy()

    out_meta.update({
        "driver": "GTiff",
        "height": out_image.shape[1],
        "width": out_image.shape[2],
        "transform": out_transform
    })

    with rasterio.open(clipped_output_file, 'w', **out_meta) as dest:
        dest.write(out_image)
```

```python
print("Raster processing and clipping completed.")
```

displaying

```python
# Install necessary packages (if not already installed)
!pip install rasterio
!pip install matplotlib

# Import necessary libraries
import rasterio
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

# Define the path to your clipped raster file
clipped_output_file = '/content/drive/MyDrive/output/2017clipped_output.tif'

# Open the clipped raster file
with rasterio.open(clipped_output_file) as src:
    # Read the raster data
    raster_data = src.read(1, masked=True)  # Read band 1 and mask nodata values

    # Set up the figure and axis
    fig, ax = plt.subplots(figsize=(10, 10))

    # Normalize the raster data to fit within the colormap range
    norm = Normalize(vmin=raster_data.min(), vmax=raster_data.max())

    # Display the raster data with the appropriate colormap and normalization
    img = ax.imshow(raster_data, cmap='viridis', norm=norm, extent=[src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top])

    # Add a colorbar
    cbar = plt.colorbar(img, ax=ax, fraction=0.03, pad=0.04)
    cbar.set_label('color')

    # Add title and labels
    plt.title('Clipped Raster Visualization')
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')

    # Show the plot
    plt.show()
```
